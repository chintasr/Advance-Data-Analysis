## Analysis 2

### 1. Can you tell me No of Vehicles by Brand Available on ebay for sale with the help of visualization

```python
df1 = df['brand'].value_counts()
plt.figure(figsize=(12, 8))
df1.plot(kind = 'bar',color = 'green',edgecolor = 'black',alpha=0.5)
plt.xticks(rotation=80)
plt.xlabel('Brand')
plt.ylabel('Number of Vehicles')
plt.title('Number of Vehicles by Brand (Bar Plot)')
plt.show()
```

### 2.What is the Average price for vehicles based on the type of vehicle as well as on the type of gearbox.Explain me with both numerical and visualization analysis¶

```python
avg_price = df.groupby(['vehicleType','gearbox'])['price'].mean()
avg_price
```

```python
avg_price = df.groupby(['vehicleType','gearbox'])['price'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='vehicleType', y='price', hue='gearbox', data = avg_price)
plt.title('Average Price for Vehicles based on Type and Gearbox')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Price')
plt.show()
```
